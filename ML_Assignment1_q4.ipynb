```python
import pandas as pd
import numpy as np

df= pd.read_csv('./WineQT.csv')
df.head(5)
```

```python
df['quality']= df['quality'].map(lambda x : 1 if x>6 else 0)
```

```python
for c in df.columns:
  if c not in ["quality"]:
    df[c]= (df[c]- df[c].min())/(df[c].max()-df[c].min())
```
